# State

```python
df.columns
```

```python
g1=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='State',ascending=False)
sns.barplot(x='State',y='Orders',data=g1)
```

```python
g1=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='State',y='Orders',data=g1)
```

***from the above graph we can observe that Uttar Pradesh has the highest number of orders.***

```python
g1=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='State',y='Amount',data=g1)
```

***from the above graphs we can conclude that the highest number of orders and sales are made by the states Uttar Pradesh, Maharashtra and Karnataka***

# Marital Status

```python
ax= sns.countplot(data=df,x='Marital_Status')

sns.set(rc={'figure.figsize':(7,5)})
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
g1=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(x='Marital_Status',y='Amount',data=g1,hue='Gender')
```

***from the above graphs we can conclude that unmarried women have a higher purchasing power***

# Occupation

```python
sns.set(rc={'figure.figsize':(20,5)})
ax= sns.countplot(data=df,x='Occupation')
for bar in ax.containers:
    ax.bar_label(bar)
```

# Product Category
